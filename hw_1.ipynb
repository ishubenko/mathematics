4. Задание (в программе):
Постройте на одном графике две кривые y(x) для функции двух переменной y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 100) # 100 точек от -5 до 5

plt.plot(x, np.cos(x * 1)) # y(x) = cos x 
plt.plot(x, np.cos(x * 3)) # y(x) = cos (x * 3) 
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```
