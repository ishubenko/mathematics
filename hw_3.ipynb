Напишите код на Python, реализующий построение графиков:

```python
r = int(input('радиус = '))
# x_0 = int(input('координаты центра окружности x_0 ='))
# y_0 = int(input('y_0 ='))

plt.figure(figsize=(6, 6))
x = np.linspace(-r, r, 1001)
y1=[]
y2=[]

for i in x:
#     j = i - y_0
    var_i_1 = math.sqrt(r**2 - i**2)
    y1.append(var_i_1)
    var_i_2 = -math.sqrt(r**2 - i**2)
    y2.append(var_i_2)

plt.plot(x,y1, color='blue')
plt.plot(x,y2, color='blue')
plt.title('Окружность')
plt.xlabel("x")
plt.ylabel("y")
```
